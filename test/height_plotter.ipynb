# Height Data Plotter

This notebook reads height samples recorded by `checkAxis.py` from `height.txt` and plots them.

## What it does
- Loads one height value per line from `height.txt`.
- Ignores empty lines.
- Builds a time axis from the sampling period.
- Plots height vs. time.

## Units
- Height values are in the Tello's own internal units (not centimeters).

## Notes
- Place this notebook in the same folder as `height.txt`, or update the file path in the code cell.
- Default sampling period is **0.1 s** (matching `checkAxis.py`).

```python
from pathlib import Path
import matplotlib.pyplot as plt

# Configuration
sampling_period_s = 0.1
data_file = Path("height.txt")

# Load data
if not data_file.exists():
    raise FileNotFoundError(
        f"Could not find {data_file.resolve()}. Run checkAxis.py first or update data_file."
    )

height_units = []
with data_file.open("r", encoding="utf-8") as file:
    for line in file:
        value = line.strip()
        if value:
            height_units.append(float(value))

if not height_units:
    raise ValueError("height.txt is empty or contains no valid numeric values.")

# Time vector
time_s = [index * sampling_period_s for index in range(len(height_units))]

# Plot
plt.figure(figsize=(10, 4))
plt.plot(time_s, height_units, linewidth=1.8)
plt.title("Tello Height Over Time")
plt.xlabel("Time (s)")
plt.ylabel("Height (Tello units)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Loaded {len(height_units)} samples from {data_file}.")
print(f"Min height: {min(height_units):.1f} Tello units")
print(f"Max height: {max(height_units):.1f} Tello units")
```
