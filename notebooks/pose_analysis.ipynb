# Pose Data Analysis

This notebook analyzes flight data from pose_log CSV files generated by `pose_plotter.py`.

## What it does
- Loads pose data (position and yaw) from CSV files
- Plots time-series data: x, y, z, yaw vs time
- Plots 3D trajectory visualization
- Computes basic statistics (min, max, mean)

## Notes
- Place CSV files in the same directory as this notebook, or update the file path in the code cell
- CSV format: `time, x, y, z, yaw` (generated by pose_plotter.py)

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from pathlib import Path
import glob

# Find all pose_log CSV files in the current directory
csv_files = sorted(glob.glob("pose_log_*.csv"))

if not csv_files:
    print("No pose_log_*.csv files found. Make sure to run pose_plotter.py first.")
    print(f"Current directory: {Path.cwd()}")
else:
    print(f"Found {len(csv_files)} CSV file(s):")
    for f in csv_files:
        print(f"  - {f}")
```

```python
# Load the most recent CSV file
if csv_files:
    data_file = csv_files[-1]  # Last file is most recent
    df = pd.read_csv(data_file)
    print(f"Loaded: {data_file}")
    print(f"\nData shape: {df.shape}")
    print(f"\nFirst few rows:")
    print(df.head())
    print(f"\nData summary:")
    print(df.describe())
```

```python
# Time-series plots
fig, axes = plt.subplots(4, 1, figsize=(12, 8))

axes[0].plot(df['time'], df['x'], linewidth=1.5, color='blue')
axes[0].set_ylabel('X (m)')
axes[0].grid(True, alpha=0.3)
axes[0].set_title('Position X vs Time')

axes[1].plot(df['time'], df['y'], linewidth=1.5, color='green')
axes[1].set_ylabel('Y (m)')
axes[1].grid(True, alpha=0.3)
axes[1].set_title('Position Y vs Time')

axes[2].plot(df['time'], df['z'], linewidth=1.5, color='red')
axes[2].set_ylabel('Z (m)')
axes[2].grid(True, alpha=0.3)
axes[2].set_title('Position Z vs Time')

axes[3].plot(df['time'], df['yaw'], linewidth=1.5, color='orange')
axes[3].set_ylabel('Yaw (rad)')
axes[3].set_xlabel('Time (s)')
axes[3].grid(True, alpha=0.3)
axes[3].set_title('Yaw Angle vs Time')

plt.tight_layout()
plt.show()
```

```python
# 3D trajectory plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(df['x'], df['y'], df['z'], linewidth=2, color='blue', label='Trajectory')
ax.scatter(df['x'].iloc[0], df['y'].iloc[0], df['z'].iloc[0], 
           color='green', s=100, label='Start', zorder=5)
ax.scatter(df['x'].iloc[-1], df['y'].iloc[-1], df['z'].iloc[-1], 
           color='red', s=100, label='End', zorder=5)

ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Z (m)')
ax.set_title('3D Trajectory')
ax.legend()
plt.show()
```

```python
# Statistics
print("=== Position Statistics ===")
print(f"\nX (m):  min={df['x'].min():.4f}, max={df['x'].max():.4f}, mean={df['x'].mean():.4f}")
print(f"Y (m):  min={df['y'].min():.4f}, max={df['y'].max():.4f}, mean={df['y'].mean():.4f}")
print(f"Z (m):  min={df['z'].min():.4f}, max={df['z'].max():.4f}, mean={df['z'].mean():.4f}")
print(f"\nYaw (rad):  min={df['yaw'].min():.4f}, max={df['yaw'].max():.4f}, mean={df['yaw'].mean():.4f}")
print(f"\nTotal flight time: {df['time'].max():.2f} seconds")
print(f"Number of samples: {len(df)}")
```
